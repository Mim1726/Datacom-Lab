```python
import matplotlib.pyplot as plt
```

```python
input = '10110010'
voltage = 1
encoded = []
```

```python
for bit in input:
    if bit=='1':
        encoded.append(voltage)
        voltage = -voltage
    else:
        encoded.append(0)
```

```python
print(encoded)
```

```python
plt.figure(figsize=(10, 3))
plt.step(range(len(encoded)), encoded, where='post')
plt.xticks(range(len(encoded)), list(input))
plt.yticks([-1, 0, 1], ["Low (-1)", "Zero (0)", "High (+1)"])
plt.xlabel("Time")
plt.ylabel("Voltage Level")
plt.title("AMI Encoding")
plt.grid(True)
plt.show()
```

```python
decoded = ""
for bit in encoded:
    if bit==0:
        decoded += '0'
    else:
        decoded += '1'
```
